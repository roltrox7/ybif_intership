```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
from sklearn.metrics import accuracy_score,precision_score,r2_score,mean_squared_error
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Fish.csv')
encoder=LabelEncoder()
df['Species']=encoder.fit_transform(df['Species'])
df.head()
```

```python
df.columns
```

```python
X=df[['Category', 'Height', 'Width', 'Length1',
       'Length2', 'Length3']]
y=df['Weight']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
